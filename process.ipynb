```python
import pandas as pd
import os
import os.path
```

```python
csv_list = []
for dirPath, dirNames, fileNames in os.walk("./dataset/"):
    for f in fileNames:
        csv_list.append(os.path.join(dirPath, f))
```

```python
df1 = pd.DataFrame()
for l_csv in range(len(csv_list)):
    df= pd.read_csv(csv_list[l_csv],",")
    df1 = pd.concat([df,df1],axis=0, ignore_index=True) 
```
