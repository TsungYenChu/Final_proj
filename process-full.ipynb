```python
import numpy as np
import pandas as pd
import os
import os.path
if not os.path.isdir("html_csv"):
    os.mkdir("html_csv")
```

```python
total= pd.read_csv("./total.csv",",")
soc = pd.read_excel("SOC.xlsx")
```

```python
al =total
al.insert(6,column='分數',value=np.nan)
```

```python
count = 0
for i in range(len(total['處理器型號'])):
    try:
        al['分數'][i]=float(soc['分數'][soc['處理器型號'].str.contains(total['處理器型號'][i], regex=False)==True].values[0])
        
    except IndexError:
        al['分數'][i]=np.nan
al
```

```python
al = al.dropna(subset=['分數'])
```

```python
al = al.sort_values(by = "分數",ascending= False)
al_low = al[al['空機價格']<=7999]
al_low.to_csv("./html_csv/score_low_decending.csv", index = False, header=True)
al_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)]
al_medium.to_csv("./html_csv/score_medium_decending.csv", index = False, header=True)
al_high = al[(al['空機價格']> 19999)]
al_high.to_csv("./html_csv/score_high_decending.csv", index = False, header=True)

# al = al.sort_values(by = "分數",ascending= True)
# al_low = al[al['空機價格']<=7999]
# al_low.to_csv("./html_csv/score_low_ascending.csv", index = False, header=True)
# al_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)]
# al_medium.to_csv("./html_csv/score_medium_ascending.csv", index = False, header=True)
# al_high = al[(al['空機價格']> 19999)]
# al_high.to_csv("./html_csv/score_high_ascending.csv", index = False, header=True)
```

```python
al = total
al = al.sort_values(by = "空機價格",ascending= False)
pr_low = al[al['空機價格']<=7999].sort_values(by = "空機價格")
pr_low.to_csv("./html_csv/price_low_decending.csv", index = False, header=True)
pr_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)]
pr_medium.to_csv("./html_csv/price_medium_decending.csv", index = False, header=True)
pr_high = al[(al['空機價格']> 19999)]
pr_high.to_csv("./html_csv/price_high_decending.csv", index = False, header=True)

al = al.sort_values(by = "空機價格",ascending= True)
pr_low = al[al['空機價格']<=7999].sort_values(by = "空機價格")
pr_low.to_csv("./html_csv/price_low_ascending.csv", index = False, header=True)
pr_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)]
pr_medium.to_csv("./html_csv/price_medium_ascending.csv", index = False, header=True)
pr_high = al[(al['空機價格']> 19999)]
pr_high.to_csv("./html_csv/price_high_ascending.csv", index = False, header=True)
```

```python
al = al.sort_values(by = "經過的月份數",ascending= True)
al.to_csv("./html_csv/month_total.csv", index = False, header=True)
al_low = al[(al['空機價格']<=7999) & (al['經過的月份數']<=12)]
al_low.to_csv("./html_csv/month_low_in.csv", index = False, header=True)
al_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)& (al['經過的月份數']<=12)]
al_medium.to_csv("./html_csv/month_medium_in.csv", index = False, header=True)
al_high = al[(al['空機價格']> 19999)& (al['經過的月份數']<=12)]
al_high.to_csv("./html_csv/month_high_in.csv", index = False, header=True)

al_low = al[(al['空機價格']<=7999) & (al['經過的月份數']>12)]
al_low.to_csv("./html_csv/month_low_out.csv", index = False, header=True)
al_medium = al[(al['空機價格']> 7999) & (al['空機價格']<=19999)& (al['經過的月份數']>12)]
al_medium.to_csv("./html_csv/month_medium_out.csv", index = False, header=True)
al_high = al[(al['空機價格']> 19999)& (al['經過的月份數']>12)]
al_high.to_csv("./html_csv/month_high_out.csv", index = False, header=True)
```
